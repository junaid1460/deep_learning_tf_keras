```python
import numpy as np
import random

def makeBin(a):
    x = "{:010b}".format(a)
    x = list(map(lambda x:x == '1', x))
    return x

def genData(a, b, evalu = False, mistake = False):
    c = a + b
    if mistake:
        c += random.randint(10,100)
    if evalu:
        print("{} + {} = {}".format(a, b, c))
    ret = [a, b, c]
    ret = list(map(makeBin, ret))
    a, b, c = ret
    return (a + b, c)
    
    
    
    
```

```python
data_x = []
data_y = []
count = 100
for i in range(count):
    a = random.randint(0, 100)
    b = random.randint(0, 100)
    x, y = genData(a, b)
    data_x.append(x)
    data_y.append(y)
```

```python
from keras.layers.core import Dense, Activation
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(100, input_shape = (20,), activation="relu"))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation = 'sigmoid'))
model.add(Dense(10, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics = ['accuracy'] )
```

```python
data_x = []
data_y = []
count = 100
for i in range(count):
    a = random.randint(0, 100)
    b = random.randint(0, 100)
    x, y = genData(a, b)
    data_x.append(x)
    data_y.append(y)

model.fit(x = data_x, y = data_y, batch_size = 100, epochs = 500)
```

```python
data_x_eval = []
data_y_eval = []
count = 1
for i in range(count):
    a = random.randint(100, 200)
    b = random.randint(0, 100)
    x, y = genData(a, b, True)
    data_x_eval.append(x)
    data_y_eval.append(y)
    
data_x_eval = np.array(data_x_eval)
data_y_eval = np.array(data_y_eval)


n, y = model.evaluate(data_x_eval, data_y_eval)
if n > y:
    print("No")
else:
    print("Yes")
```
