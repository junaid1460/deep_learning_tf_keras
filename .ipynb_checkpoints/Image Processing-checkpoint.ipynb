# Libraries

```python
import PIL.Image as image
import PIL.ImageOps as ops
import matplotlib.pyplot as plt
```

```python
img = image.open('dd.jpg')
img = ops.grayscale(img)
```

# Thresholding
here I process pixel values from 0 to 255 and mapping to thresh function

```python
def thresh(x):
    if x > 125 and x < 150:
        return x
    else:
        return 0
x = img.point(lambda x: thresh(x))
x
```

# Segmentation
Here all the pixels are biased towards 0 and 255 based on threshold

```python
def segment(x):
    if x < 195:
        return 0;
    else:
        return 255
x = img.point(lambda x: segment(x))
x
```

# Histogram

```python
plt.plot(x.histogram())
plt.show()
```

```python
# saving in other id for future usage
img = x
oldimg = x
```

```python
import PIL.ImageFilter as imf
#Image filter for creating filter kernels
img = image.open('dd.jpg')
```

# Find all edges

```python
img.filter(imf.FIND_EDGES)
```

# Edge detection with custom kernels

```python
top = imf.Kernel(kernel=[0,1,0,1,0,0,0,0,0], size=(3,3)) # roberts
top = imf.Kernel(kernel=[-1,0,1,-2,0,2,-1,0,1], size=(3,3)) #
top = imf.Kernel(kernel=[-1,-2,-1,0,0,0,1,2,1], size=(3,3))


# apply filter
oldimg.filter(filter=top)
```
